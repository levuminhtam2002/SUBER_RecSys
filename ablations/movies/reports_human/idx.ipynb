```python
import pandas as pd
import numpy as np
import csv
df_template = pd.read_csv("./TheBloke_Llama-2-13B-chat-GPTQ/samples.csv")


idx = range(len(df_template)*5)

rng = np.random.default_rng(42)
idx = rng.permutation(idx)
```

```python
folders = ["gpt-4-0613", "TheBloke_Llama-2-13B-chat-GPTQ", "TheBloke_Mistral-7B-Instruct-v0.2-GPTQ", "TheBloke_vicuna-13B-v1.5-GPTQ",
           "TheBloke_vicuna-13B-v1.5-GPTQ-random_rating"]

counter = 0
for f in folders:
    df = pd.read_csv(f"./{f}/samples.csv")
    df["interaction_id"] = idx[counter:counter+len(df)]
    counter += len(df)
    df = df.drop(columns=["Unnamed: 0"])
    df.to_csv(f"./{f}/samples_form.csv", quoting=csv.QUOTE_ALL)
```
